```python
# BLACKJACK GAME

import random

def deal_card():
    """
    return a random card from the deck
    
    """
    cards=[11,2,3,4,5,6,7,8,9,10,10,10,10]
    card=random.choice(cards)
    return card


def compare(u_score,c_score):
    if u_score == c_score:
        return "Draw"
    elif c_score==0:
        return "Lose"
    elif u_score==0:
        return "win with a blackjack"
    elif u_score > 21:
        return "you went over. you lose"
    elif c_score > 21:
        return "opponent went ower. you win"
    elif u_score > c_score:
        return "you win"
    else:
        return "you lose"



def play_game():
    user_cards=[]
    computer_cards=[]
    computer_score=-1
    user_score=-1
    is_game_over=False

    for _ in range(2):
        user_cards.append(deal_card())
        computer_cards.append(deal_card())


    def calculate_score(cards):
        """
        take a list of cards and calculate scare od cards

        """
        if sum(cards)==21 & len(cards)==2:
            return 0
        if 11 in cards and sum(cards)>21:
            cards.remove(11)
            cards.append(1)

        return sum(cards)  

    while not is_game_over:
        user_score=calculate_score(user_cards)  

        computer_score=calculate_score(computer_cards)  


        print(f"your cards: {user_cards}, current score: {user_score}")
        print(f"computers first card: {computer_cards[0]}")

        if user_score==0 or computer_score==0 or user_score>21 :
            is_game_over=True

        else:
            user_should_deal=input("type 'y' to get another card, type 'n' to pass: ")
            if user_should_deal=="y":
                user_cards.append(deal_card())

            else:
                is_game_over=True    


    while computer_score !=0 and computer_score < 17 :
        computer_cards.append(deal_card())
        computer_score=calculate_score(computer_cards)

    print(f"your final hand: {user_cards} and your score: {user_score}")
    print(f"computers final hand: {computer_cards} and its score: {computer_score}")

    print(compare(user_score,computer_score))


while input("Do you want to play a game of BLACHJACK? 'y' or 'n' ")=="y":
    play_game()
```

```python
import time

def dely_decorator(function):
                                           # nsted func: wrapper_func()
    def wrapper_func():   
        time.sleep(2)
                                           # do sth before func
        function()
                                           # do sth after func
    return wrapper_func  


@dely_decorator
def say_hello():                          # dely_decorator(say_hello())  -->> 1. do befor   2. do main func(say_hello())  3. do after
    print("hello")


@dely_decorator
def say_Bye():
    print("Bye")


def say_greeting():
    print("How are you?")    


#result
say_hello()          # with delay
say_Bye()            # with delay
say_greeting()       # without delay
```

```python
class User:
    def __init__(self,name):
        self.name=name
        self.is_logged_in=False


def is_authenticated_decorator(function):
    def wrapper(*args,**kwrgs):

        if args[0].is_logged_in==True:
            function(args[0])
            
    return wrapper            



@is_authenticated_decorator
def create_blog_post(user):
    print(f"This is {user.name} 's new blog post. ")


new_user=User("yekta")
new_user.is_logged_in=True
create_blog_post(new_user)
```

```python
def decorator(function):
    def wrapper(*args):
        assert len(args)>=2, "wrong input"
        acc = function(args[0],args[1])
        for a in args[2:]:
            acc=function(acc, a)
        return acc
    return wrapper




@decorator
def concate(s1,s2):
    return f"{s1}{s2}"

@decorator
def multiple(a,b):
    return a*b

print(concate("a","a","a","a","a"))
print(multiple(1,2,3))
```

```python
def logging_decorator(function):
    def wrapper(*args):
        res=function(*args)
        return res
    return wrapper    




@logging_decorator
def a_function(*args):
    return sum(args)


a_function(1,2,3,4,5,6)
```

```python

def extract_text_from_document(file):
    import os
    extension = os.path.splitext(file)

    match extension:        
        case '.pdf':
            from langchain.document_loaders import PyPDFLoader
            print(f'Loading {file}')
            loader = PyPDFLoader(file)

        case '.docx':
            from langchain.document_loaders import Docx2txtLoader
            print(f'Loading {file}')
            loader = Docx2txtLoader(file)

        case None:
            print('Document format is not supported by our application!')
            return None

    data = loader.load()
    return data
```

```python
import pandas as pd
import sqlite3
from faker import Faker  

## Connect DB
conn = sqlite3.connect("/home/yekta/new_lang/lang_new/data.sqlite")
cursor = conn.cursor()
tables = cursor.execute(f"SELECT * FROM 'supermarket_sales - Sheet1';")
all_data=tables.fetchall()
column_names = [description[0] for description in cursor.description]
conn.close

# print(column_names)

df=pd.DataFrame(data=all_data, columns=column_names)
print(f"====YOUR DATABASE====: \n {'supermarket_sales - Sheet1'} \n{df}")


## convert En to Fa:df['City']
unique_values = df['City'].unique()

fake=Faker('fa_IR')
unique_cities=df['City'].unique()
print(unique_cities)

fake_city_map = {city: fake.state() for city in unique_cities}
assert  len(fake_city_map)==len(set(fake_city_map.values()))
df['City'] = df['City'].map(fake_city_map)


## convert En to Fa:df['Product line']
df['Product line'].unique()
fake_product_map={'Health and beauty':' ',
                   'Electronic accessories':"  ",
                   'Home and lifestyle':"   ",
                   'Sports and travel':"  ",
                   'Food and beverages':"  ",
                    'Fashion accessories':" "}
df['Product line'] = df['Product line'].map(fake_product_map)


## convert En to Fa:df['Gender']
df['Gender'].unique()
gender_map={'Female':'',
                   'Male':"",
                 }
df['Gender'] = df['Gender'].map(gender_map)


## convert En to Fa:df['Customer type']
df['Customer type'].unique()
Customer_type_map={'Member':'',
                   'Normal':"",
                   }
df['Customer type'] = df['Customer type'].map(Customer_type_map)


## convert En to Fa:df['Payment']
df['Payment'].unique()
Payment_map={'Ewallet':'  ',
                   'Cash':" ",
                   'Credit card':" "
                   }
df['Payment'] = df['Payment'].map(Payment_map)


## convert En to Fa:df['Branch']
df['Branch'].unique()
Branch_map={'A':'',
            'B':"",
            'C':""
                   }
df['Branch'] = df['Branch'].map(Branch_map)


## converted  df.Total
df["Total"] = df["Total"].astype(float).round().astype(int)
df["Total"] = [row * 1000 for row in df["Total"]]

## converted  df.gross income
df["gross income"] = df["gross income"].astype(float).round().astype(int)
df["gross income"] = [row * 100 for row in df["gross income"]]

## converted  df.Unit price
df["Unit price"] = df["Unit price"].astype(float).round().astype(int)
df["Unit price"] = [row * 10000 for row in df["Unit price"]]

## converted  df.Date.sort
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values(by="Date")
print(df.Date[:])

## converted DataFrame
print(f"====NEW DATABASE that converted to Persian====: \n{df}")

# Save the DataFrame to an SQLite table called 'city_data'
conn = sqlite3.connect('supermarket_sales - Sheet1')
df.to_sql('DataBase.sqlite', conn, if_exists='replace', index=False)
print(f"saved your data in sqlite3 db with name=DataBase.sqlite")
```
